```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def ages(mu=50,sig=13,n=100,seed=42):
    
    np.random.seed(seed)

    samples = np.random.normal(loc=mu,scale=sig,size=n)
    samples = np.round(samples,decimals=0)

    return samples
```

```python
sample = ages()
```

```python
sample
```

```python
sns.distplot(sample)
sns.set_style("darkgrid")
```

```python
sns.boxplot(sample)
```

```python
ser = pd.Series(sample)
ser.describe()
```

```python
ser[ser>18]
```

```python
def nan(df):
    vals = df.isnull().sum()/len(df)*100
    vals = vals[vals>0].sort_values()
    
    return vals
```

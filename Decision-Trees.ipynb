```python
X = pd.get_dummies(df.drop('species',axis=1),drop_first=True)
Y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(Xtrain,Ytrain)
```

```python
preds = tree.predict(Xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(Ytest,preds)
```

```python
print(classification_report(Ytest,preds))
```

```python
tree.feature_importances_
```

```python
pd.DataFrame(index=X.columns,data=tree.feature_importances_)
```

```python
sns.boxplot(data=df,x='body_mass_g',y='species')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(14,10))
plot_tree(tree,filled=True,feature_names=X.columns)
```

```python
def report_model(model):
    model_preds = model.predict(Xtest)
    print(classification_report(Ytest,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=X.columns);
```

```python
pruned  = DecisionTreeClassifier(max_depth=2)
```

```python
pruned.fit(Xtrain,Ytrain)
```

```python
report_model(pruned)
```

```python
tree2 = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
tree2.fit(Xtrain,Ytrain)
```

```python
report_model(tree2)
```

```python
entropy = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy.fit(Xtrain,Ytrain)
```

```python
report_model(entropy)
```

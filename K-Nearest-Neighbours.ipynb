```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../DATA/sonar.all-data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.corr())
```

```python
df.corr()
```

```python
df['target'] = df['Label'].map({'R':0,'M':1})
```

```python
abs(df.corr()['target']).sort_values().tail(6)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Label','target'],axis=1)
Y = df['Label']
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y,test_size=0.1,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()

ops = [('scaler',scaler),('knn',knn)]
```

```python
from sklearn.pipeline import Pipeline
pipe  = Pipeline(ops)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
kvals = list(range(1,30))
params = {'knn__n_neighbors':kvals}
```

```python
fullCV = GridSearchCV(pipe,param_grid= params,cv=5,scoring='accuracy')
```

```python
fullCV.fit(Xtrain,Ytrain)
```

```python
fullCV.best_estimator_.get_params()
```

```python
jj =fullCV.cv_results_['mean_test_score']
```

```python
plt.plot(jj)
```

```python
preds= fullCV.predict(Xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(Ytest,preds)
```

```python
print(classification_report(Ytest,preds))
```
